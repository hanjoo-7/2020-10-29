# 4일째 되는 날 프로그램 매수 등락률

```python
%matplotlib inline
import matplotlib.pyplot as plt
# unicode minus를 사용하지 않기 위한 설정 (minus 깨짐현상 방지)
plt.rcParams['axes.unicode_minus'] = False
plt.rcParams['font.family'] = 'Malgun Gothic'
import pandas as pd
import seaborn as sns
```

```python
투자자 = pd.read_csv('투자자20201103.csv', encoding='euc-kr')
```

```python
프로그램매수 = 투자자[
    (투자자.프로그램순매수1>0) &
    (투자자.프로그램순매수2>0) &
    (투자자.프로그램순매수3>0)&
    (투자자.프로그램순매수4>0)]

#4일 매수 이후 수익률
프로그램매수['등락률4매수10매도'] = (프로그램매수['종가10'] - 프로그램매수['종가4']) / 프로그램매수['종가4']
프로그램매수['등락률4매수9매도'] = (프로그램매수['종가9'] - 프로그램매수['종가4']) / 프로그램매수['종가4']
프로그램매수['등락률4매수8매도'] = (프로그램매수['종가8'] - 프로그램매수['종가4']) / 프로그램매수['종가4']
프로그램매수['등락률4매수7매도'] = (프로그램매수['종가7'] - 프로그램매수['종가4']) / 프로그램매수['종가4']
프로그램매수['등락률4매수6매도'] = (프로그램매수['종가6'] - 프로그램매수['종가4']) / 프로그램매수['종가4']
프로그램매수['등락률4매수5매도'] = (프로그램매수['종가5'] - 프로그램매수['종가4']) / 프로그램매수['종가4']

프로그램매수.head()
```

```python
프로그램_4매수_등락률 = 프로그램매수.groupby('종목').agg(
    {
        '등락률4매수10매도':'mean',
        '등락률4매수9매도':'mean',
        '등락률4매수8매도':'mean',
        '등락률4매수7매도':'mean',
        '등락률4매수6매도':'mean',
        '등락률4매수5매도':'mean'
    }
)

프로그램_4매수_등락률
```

```python
#프로그램_4매수_등락률 = 프로그램_4매수_등락률.sort_values('등락률4매수9매도', ascending=False)
```

## 매도 날짜에 따른 평균 등락률

```python
average = pd.DataFrame(columns = ['매수/매도','평균 등락률'])

for each in 프로그램_4매수_등락률.columns:
    print(each ,프로그램_4매수_등락률[each].mean())
    average = average.append(pd.DataFrame([[each, 프로그램_4매수_등락률[each].mean()]],columns = ['매수/매도','평균 등락률']),ignore_index = True)
```

```python
#최대 평균 등락률
average['평균 등락률'].max()
```
